### üîπ HTTP y JSON (D√≠a 1)

#### 1-Crear un diccionario en Python y convertirlo en JSON (json.dumps).

#### 2-Leer un JSON desde string y acceder a sus claves (json.loads).

#### 3-Hacer un GET a una API p√∫blica con requests.get.

#### 4-Hacer un POST enviando un JSON a un servidor de prueba (ej: httpbin.org/post).

#### 5-Simular un header Authorization en un request con requests.

```python
# 1-Crear un diccionario en Python y convertirlo en JSON (json.dumps).
import json

persona = {
    "Nombre": "Efra",
    "Edad": 20,
    "Ciudad": "Asuncion"
}

persona_json = json.dumps(persona)
print(persona_json)
```

```python
# 2-Leer un JSON desde string y acceder a sus claves (json.loads).
persona_1 = json.loads(persona_json)
print(persona_1["Nombre"])
print(persona_1["Edad"])
```

```python
# 3-Hacer un GET a una API p√∫blica con requests.get.

import requests

response = requests.get("https://jsonplaceholder.typicode.com/users")
data = response.json()
print(data[0])
```

```python
# 4-Hacer un POST enviando un JSON a un servidor de prueba (ej: httpbin.org/post).

import requests 
import json

persona_3 = {
    "Nombre": "Juan",
    "Edad": 25,
    "Ciudad": "Asuncion"
}

response = requests.post("https://httpbin.org/post", json = persona_3)
print(response.json())
```

```python
# 5-Simular un header Authorization en un request con requests.

import requests

headers = {"Authorization": "Bearer mi_token_123"}

response = requests.get("https://httpbin.org/headers", headers=headers)
print(response.json())
```

### üîπ Flask / FastAPI (D√≠a 1‚Äì2)

#### 1-Crear un servidor Flask con un endpoint /ping que responda "pong".

#### 2-Hacer un endpoint /echo que devuelva exactamente el JSON que recibe.

#### 3-Validar un header: si no existe Authorization, devolver error 401.

#### 4-Crear una lista de tokens v√°lidos y validar que el recibido est√© en esa lista.

#### 5-Crear un endpoint que reciba un JSON con {name, age} y devuelva "Hola, name".

```python
# 1-Crear un servidor Flask con un endpoint /ping que responda "pong".

from flask import Flask

app = Flask(__name__)

@app.route("/ping")
def ping():
    return "pong"

if __name__ == "__main__":
    app.run(debug=True, use_reloader=False)
```

```python
# 2-Hacer un endpoint /echo que devuelva exactamente el JSON que recibe.
from flask import Flask, request, jsonify

app = Flask(__name__)

@app.route("/echo", methods = ["POST"])
def echo():
    data = request.get_json()
    return jsonify(data)

if __name__ == "__main__":
    app.run(debug=True, use_reloader=False)
```

```python
import requests

url = "http://127.0.0.1:5000/echo"
payload = {"mensaje": "hola"}

response = requests.post(url, json=payload)
print(response.json())
```
